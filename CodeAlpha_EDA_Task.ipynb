## **STEP 4.4 : Detect Anomalies (Outliers)**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Sales'])
plt.title('Sales Outliers')
plt.show()
```

###**Use Method: IQR Method**

```python
# Calculate Q1 and Q3
Q1 = data['Sales'].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)

# IQR
IQR = Q3 - Q1

# Lower and Upper limit
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)
```

```python
# Remove outliers
data = data[(data['Sales'] >= lower_limit) &
            (data['Sales'] <= upper_limit)]

print("New Shape:", data.shape)
```

###**Check Again With Boxplot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Sales'],color='pink')
plt.xlabel('Sales')
plt.title('Sales After Removing Outliers')
plt.show()
```

# **Observation**

* The Sales column initially contained several extreme values (outliers).
* Boxplot visualization was used to identify abnormal data points.
* After handling outliers, the data distribution became more balanced and cleaner.

---

# **Action Taken**

* The **IQR (Interquartile Range)** method was applied to detect and handle outliers.
* Extreme Sales values outside the acceptable range were removed/capped to reduce data distortion.

---

# **Insights**

* The dataset became cleaner and more suitable for analysis.
* Outlier treatment improved overall data quality.
* Future analysis and predictive modeling accuracy can improve with reduced noise in the dataset.
* Sales distribution is now more stable and easier to interpret visually.

---

# **Final Conclusion**

* Outlier handling is an important step in Exploratory Data Analysis (EDA).
* Removing or controlling extreme values helps produce more reliable insights.
* Properly cleaned data improves business analysis, visualization quality, and model performance.


```python
# Capping outliers
data['Sales'] = np.where(data['Sales'] > upper_limit,
                         upper_limit,
                         data['Sales'])

data['Sales'] = np.where(data['Sales'] < lower_limit,
                         lower_limit,
                         data['Sales'])
```

###**Step 4.5: Correlation Pattern**

```python
# ==============================
# Correlation Heatmap
# ==============================

import matplotlib.pyplot as plt
import seaborn as sns

# Select numeric columns
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Create correlation matrix
corr_matrix = numeric_data.corr()

# Plot Heatmap
plt.figure(figsize=(10,7))

sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Correlation Heatmap')

plt.show()
```

# **Insights**

* **Sales and Profit** have a **strong positive correlation (0.87)**, meaning higher sales generally lead to higher profit.
* **Sales and Unit Price** also show a strong positive relationship (0.65).
* **Unit Price and Profit** are positively correlated (0.55), indicating that higher-priced products contribute more profit.
* **Quantity and Unit Price** have a moderate negative correlation (-0.38), suggesting higher quantity sales are often linked with lower-priced products.
* **Discount** has a very weak negative correlation with Profit (-0.08), indicating discounts slightly reduce profitability.
* Columns like **Row ID** and **Postal Code** have almost no meaningful relationship with business performance metrics.

---

# **Final Conclusion**

* Sales growth plays a major role in improving overall profitability.
* Higher unit-priced products contribute positively to both Sales and Profit.
* Excessive discounts should be controlled to protect profit margins.
* Quantity alone does not strongly impact Profit without proper pricing strategies.
* Correlation analysis helps identify the most influential business variables for decision-making and future predictive analysis.

```python
# ==============================
# Group Analysis
# ==============================

import matplotlib.pyplot as plt
import seaborn as sns

# Group data
sales_category = data.groupby('Category')['Sales'].sum().reset_index()
profit_category = data.groupby('Category')['Profit'].sum().reset_index()

# Create figure
fig, axes = plt.subplots(2, 1, figsize=(8,10))

# ------------------------------
# Total Sales by Category
# ------------------------------
sns.barplot(x='Category',
            y='Sales',
            data=sales_category,
            palette='Set2',
            ax=axes[0])

axes[0].set_title('Total Sales by Category')
axes[0].set_ylabel('Total Sales')

# Add values on bars
for i, v in enumerate(sales_category['Sales']):
    axes[0].text(i, v + 1000, f'{v:,.0f}', ha='center')

# ------------------------------
# Total Profit by Category
# ------------------------------
sns.barplot(x='Category',
            y='Profit',
            data=profit_category,
            palette='Set2',
            ax=axes[1])

axes[1].set_title('Total Profit by Category')
axes[1].set_ylabel('Total Profit')

# Add values on bars
for i, v in enumerate(profit_category['Profit']):
    axes[1].text(i, v + 500, f'{v:,.0f}', ha='center')

# Layout
plt.tight_layout()

# Show plots
plt.show()
```
